```python
%%capture
!pip install nltk
import nltk
nltk.download("words")
from nltk.corpus import words
import numpy as np
import json
import pylab as plt
```

```python
character_count = 0
letter_count = {}
alphabet = "abcdefghijklmnopqrstuvwxyz-"

for letter in alphabet:
    letter_count[letter] = 0

for word in words.words():
    for letter in word.lower():
        letter_count[letter] += 1
    character_count += len(word)
    
for k, v in letter_count.items():
    letter_count[k] = v / character_count
    
plt.plot(np.array(list(letter_count.values())))
plt.show()
```

```python
with open("english_letter_frequency.json", "w") as fd:
    json.dump(letter_count, fd)
```
